```python
# YOUR CODE
# YOUR CODE HERE
carbon = fuel_econ['co2']
range = np.arange(100, 800, 100)

plt.hist(carbon, bins=range, edgecolor='black', label='CO2 Emission')
plt.xlabel('CO2')
plt.ylabel('Frequency')
plt.title("CO2 Distribution")
plt.savefig('CO2_destribution.png')
plt.show

# The distribution is symmetric because data distribution is roughly concentrated around the center
```

```python
# YOUR CODE HERE
plt.hist(fuel_econ.trans, orientation='horizontal', edgecolor='gray')
plt.xlabel("Transmission")
plt.ylabel("Frequency")
plt.title('Distribution of transmissiontypes')
plt.savefig('transmissiontypes.png')
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(fuel_econ.fuelType, orientation='horizontal', edgecolor='gray')
plt.xlabel("Frequency")
plt.ylabel("Fuel Type")
plt.title('Fuel Types by Usage')
plt.savefig('fuel_types.png')
plt.show()
```
